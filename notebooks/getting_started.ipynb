# MLOps Monitoring Project - Getting Started

This notebook demonstrates how to use the MLOps monitoring framework.

```python
import sys
sys.path.append('../')

from src.training.train import ModelTrainer
from src.inference.predict import ModelInference
from src.monitoring.drift_monitor import DataDriftMonitor
from config.config import config

import pandas as pd
import numpy as np
```

## 1. Generate Sample Data

```python
# Generate sample dataset
np.random.seed(42)
n_samples = 1000

data = {
    'feature_1': np.random.normal(0, 1, n_samples),
    'feature_2': np.random.normal(2, 1.5, n_samples),
    'feature_3': np.random.uniform(-1, 1, n_samples),
    'target': np.random.choice([0, 1], n_samples)
}

df = pd.DataFrame(data)
df.to_csv('../data/raw/sample_data.csv', index=False)
print(f"Generated {len(df)} samples")
df.head()
```

## 2. Train Model

```python
trainer = ModelTrainer()
data = trainer.load_data('../data/raw/sample_data.csv')
X_train, X_test, y_train, y_test = trainer.preprocess_data(data, 'target')

model = trainer.train_model(X_train, y_train)
metrics = trainer.evaluate_model(X_test, y_test)

print(f"Model Accuracy: {metrics['accuracy']:.3f}")
trainer.save_model('../models/sample_model.pkl')
```

## 3. Monitor Data Drift

```python
# Create reference data (training data)
reference_data = X_train

# Simulate new data with drift
new_data = X_test.copy()
new_data['feature_1'] = new_data['feature_1'] + 0.5  # Introduce drift

# Initialize drift monitor
drift_monitor = DataDriftMonitor(reference_data)
drift_report = drift_monitor.detect_drift(new_data)

print("Drift detection completed. Check the report for details.")
```
