# Fraud Detection Model Training
Train and deploy fraud detection model with SageMaker

```python
import sagemaker
from sagemaker.sklearn import SKLearn
from sagemaker.inputs import TrainingInput
import boto3
```

```python
# Setup
session = sagemaker.Session()
role = 'arn:aws:iam::YOUR_ACCOUNT:role/MLOpsMonitoringRole'  # Replace
bucket = 'your-mlops-monitoring-bucket'  # Replace

print(f"SageMaker role: {role}")
print(f"S3 bucket: {bucket}")
```

```python
# Create SKLearn estimator
sklearn_estimator = SKLearn(
    entry_point='fraud_train.py',
    source_dir='../src/training',
    role=role,
    instance_type='ml.m5.large',
    framework_version='0.23-1',
    py_version='py3',
    hyperparameters={
        'n_estimators': 100,
        'max_depth': 10
    }
)
```

```python
# Train model
train_input = TrainingInput(
    s3_data=f's3://{bucket}/data/fraud_dataset.csv',
    content_type='text/csv'
)

sklearn_estimator.fit({'train': train_input})

print(f"Model artifacts: {sklearn_estimator.model_data}")
```
